```python
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,2*np.pi,100)
Ya = np.sin(X)
Yb = np.cos(X)
```

```python
plt.plot(X,Ya)
plt.plot(X,Yb)
plt.show()
```

The first two arguments in `np.linspace()` denote the start and end of the range of values. The final argument specifies how many partitions to break the range into.

```python
plt.plot(X,Ya)
plt.plot(X,Yb)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
